```python
from PIL import Image

# 读取图片
image = Image.open('D:\GitHub\BigData\Visualization\Tasks\Task6\src\PremierLeague\premier-league-bix-six.webp')

# 将图片转换为带有透明通道的RGBA模式
image = image.convert('RGBA')

# 获取图片的像素数据
pixels = image.getdata()

# 降低透明度
new_pixels = []
for pixel in pixels:
    r, g, b, a = pixel
    new_pixels.append((r, g, b, int(a * 0.1)))  # 降低透明度为原来的一半

# 更新图片的像素数据
image.putdata(new_pixels)

# 保存修改后的图片
image.save('D:\GitHub\BigData\Visualization\Tasks\Task6\src\PremierLeague\BIG6_with_reduced_opacity.png')
```
